```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df1['Sex']=le.fit_transform(df1['Sex'])
df1['ChestPainType']=le.fit_transform(df1['ChestPainType'])
df1['RestingECG']=le.fit_transform(df1['RestingECG'])
df1['ExerciseAngina']=le.fit_transform(df1['ExerciseAngina'])
df1['ST_Slope']=le.fit_transform(df1['ST_Slope'])
```

```python
df1.head()
```

```python
x=df1.drop(['HeartDisease'],axis='columns')
y=df1.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaler=scaler.fit_transform(x)
x_scaler
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaler,y,test_size=0.2,random_state=20)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
scores = cross_val_score(SVC(), x, y, cv=5)
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
bag_model = BaggingClassifier(base_estimator=SVC(), n_estimators=100, max_samples=0.8, random_state=0)
scores = cross_val_score(bag_model, x, y, cv=5)
scores.mean()
```

```python
from sklearn.tree import DecisionTreeClassifier
scores = cross_val_score(DecisionTreeClassifier(random_state=0), x, y, cv=5)
scores.mean()
```

```python
bag_model = BaggingClassifier(
    base_estimator=DecisionTreeClassifier(random_state=0),
    n_estimators=100,
    max_samples=0.9,
    oob_score=True,
    random_state=0
)
scores = cross_val_score(bag_model, x, y, cv=5)
scores.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
scores = cross_val_score(RandomForestClassifier(), x, y, cv=5)
scores.mean()
```
