```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
dir(digits)
```

```python
digits
```

```python
digits.data[0]
```

```python
plt.gray()
plt.matshow(digits.images[0])
```

```python
plt.gray()
for i in range(5):
  plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
plt.matshow(digits.images[36])
```

```python
digits.target[36]
```

```python
model.predict([digits.data[36]])
```

```python
digits.target[67]
```

```python
model.predict([digits.data[67]])
```

```python
y_predict=model.predict(X_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_predict)
cm
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(cm,annot=True)
plt.title("digits data")
plt.xlabel("predicted")
plt.ylabel("orginal")
```

```python
from sklearn.preprocessing import LabelEncoder
la=LabelEncoder()
```

```python
df['Species']=la.fit_transform(df['Species'])
```

```python
df
```

```python
x=df.drop(["Species","Id"],axis="columns")
x
```

```python
y=df.Species
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
X_train
```

```python
len(X_train)
```

```python
X_test
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
Y_predict=reg.predict(X_test)
Y_predict
```

```python
from sklearn.metrics import confusion_matrix
rm=confusion_matrix(y_test,Y_predict)
rm
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(rm,annot=True)
plt.title("Iris data")
plt.xlabel("predicted")
plt.ylabel("orginal flower")
```
