```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_digits
dg=load_digits()
```

```python
dir(dg)
```

```python
for i in range(6):
  plt.matshow(dg.images[i])
```

```python
df=pd.DataFrame(dg.data,columns=dg.feature_names)
```

```python
df['target']=dg['target']
```

```python
df.head()
```

```python
df.describe()
```

```python
x=df.drop(['target'],axis="columns")
x
```

```python
y=df.target
y
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
x_scaled=ss.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.3,random_state=20)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)
x_pca=pca.fit_transform(x)
x_pca
```

```python
x_pca.shape
```

```python
pca.n_components_
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_pca,y,test_size=0.2,random_state=30)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x)
X_pca.shape
```

```python
X_pca
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_pca,y,test_size=0.2,random_state=20)
```

```python
len(x_train)
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df1['Sex']=le.fit_transform(df1['Sex'])
df1['ChestPainType']=le.fit_transform(df1['ChestPainType'])
df1['RestingECG']=le.fit_transform(df1['RestingECG'])
df1['ExerciseAngina']=le.fit_transform(df1['ExerciseAngina'])
df1['ST_Slope']=le.fit_transform(df1['ST_Slope'])
```

```python
df1.head()
```

```python
x=df1.drop(['HeartDisease'],axis='columns')
y=df1.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=30)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
from sklearn.decomposition import PCA

pca = PCA(0.95)
X_pca = pca.fit_transform(x)
X_pca
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_pca,y,test_size=0.2,random_state=30)
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```
