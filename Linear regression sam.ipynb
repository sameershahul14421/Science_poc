```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
r=pd.read_csv("hiring.csv")
r
```

```python
r.experience = r.experience.fillna("zero")
r
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
r['experience']=le.fit_transform(r['experience'])
r
```

```python
import math
mean_test_score=math.floor(r["test_score(out of 10)"].mean())
mean_test_score
```

```python
r["test_score(out of 10)"]=r["test_score(out of 10)"].fillna(mean_test_score)
r
```

```python
reg=linear_model.LinearRegression()
reg.fit(r[["experience","test_score(out of 10)","interview_score(out of 10)"]],r[["salary($)"]])
```

```python
reg.predict([[2,9,6]])
```

```python
reg.predict([[12,10,10]])
```
